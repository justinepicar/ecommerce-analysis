```python
avg_before_purchase='''SELECT productSKU, ( sum_of_hit_number / total_hits ) AS avg_hit_number
                       FROM ( 
                              SELECT productSKU, SUM(hitNumber) AS sum_of_hit_number
                              FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20160801`,
                              UNNEST(ARRAY(totals)),
                              UNNEST(hits),
                              UNNEST(hits.product) product
                              WHERE productSKU IS NOT NULL
                              AND productSKU=GGOEGESB015199
                              AND totals.transactions>=1
                              GROUP BY productSKU 
                             ) AS product1
                       JOIN (
                               SELECT productSKU, COUNT( fullVisitorId ) AS total_hits
                               FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20160801`,
                               UNNEST(ARRAY(totals)),
                               UNNEST(hits),
                               UNNEST(hits.product) product
                               WHERE productSKU IS NOT NULL
                               AND productSKU=GGOEGBFC018799
                               AND totals.transactions>=1
                               GROUP BY productSKU 
                             ) AS product2
                       ON product1.productSKU = product2.productSKU;'''
#hits_product_df = client.query(avg_before_purchase).to_dataframe()
#hits_product_df.head()
```

```python
query= '''
WITH ga_tables AS (
SELECT
date,
SUM(totals.visits) AS visits,
SUM(totals.pageviews) AS pageviews,
SUM(totals.transactions) AS transactions,
SUM(totals.transactionRevenue)/1000000 AS revenue
FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20160801`
GROUP BY date

UNION ALL

SELECT
date,
SUM(totals.visits) AS visits,
SUM(totals.pageviews) AS pageviews,
SUM(totals.transactions) AS transactions,
SUM(totals.transactionRevenue)/1000000 AS revenue
FROM `bigquery-public-data.google_analytics_sample.ga_sessions_20160802`
GROUP BY date

)
SELECT
date,
visits,
pageviews,
transactions,
revenue,
FROM ga_tables
ORDER BY date ASC'''
client.query(query).to_dataframe()
```

```python
def query_nested_hits(key):
    
    query=f'''SELECT * FROM UNNEST(ARRAY(
                 SELECT {key}
                 FROM `bigquery-public-data.google_analytics_sample.ga_sessions_*`,
                 UNNEST(hits)
                 WHERE _TABLE_SUFFIX BETWEEN '20170801' AND '20170801'
                 LIMIT 1000))
                '''
    return client.query(query).to_dataframe()
```

```python
#What is the total number of transactions generated per device or browser from June 2017 to August 2017?
query='''SELECT * FROM (
                        SELECT IFNULL(SUM(totals.transactions),0) AS Transactions , device.browser AS device
                        FROM `bigquery-public-data.google_analytics_sample.ga_sessions_*`
                        WHERE _table_suffix BETWEEN '20170601' AND '20170830'
                        GROUP BY device 
                        ORDER BY Transactions DESC
                        )
         WHERE Transactions > 0'''

client.query(query).to_dataframe()
```

```python
# PAGE VIEWS

#What is the sequence of pages viewed?

query = ''' SELECT fullVisitorId, 
                   visitId, 
                   visitNumber,
                   hits.hitNumber AS hitNumber,
                   hits.page.pagePath AS pagePath,
                   IFNULL(totals.transactions, 0) AS transactions
            FROM `bigquery-public-data.google_analytics_sample.ga_sessions_*`,
                  UNNEST(hits) as hits
            WHERE _TABLE_SUFFIX BETWEEN '20170601' AND '20170830'
                  AND hits.type="PAGE"
            ORDER BY fullVisitorId,
                     visitId,
                     visitNumber,
                     hitNumber '''
sequencePath_df = client.query(query).to_dataframe()

sequencePath_df.to_csv('sequencePath_20170601_to_20170830')
#sequencePath_df = pd.read_csv('sequencePath_20170601_to_20170830')

sequencePath_df[sequencePath_df.transactions > 0].head(10)
```

```python
# What was the total number of pageviews, visits, transactions, and revenue from June 2017 to August 2017?
query= '''SELECT date,
                 SUM(totals.visits) AS visits,
                 SUM(totals.pageviews) AS pageviews,
                 SUM(totals.transactions) AS transactions,
                 SUM(totals.transactionRevenue)/1000000 AS revenue_millions
          FROM `bigquery-public-data.google_analytics_sample.ga_sessions_*`
          WHERE _TABLE_SUFFIX BETWEEN '20170601' AND '20170830'
          GROUP BY date
          ORDER BY date ASC;'''

total_views = client.query(query).to_dataframe()
total_view.head()

#What was the average number of product pageviews per day for users who made a purchase in June/July/August 2017?

query = ''' SELECT (SUM(pageviews)/SUM(users)) AS avg_number_of_product_pageviews
            FROM (
                 SELECT sum(totals.pageviews) as pageviews,
                   COUNT(fullvisitorid) as users
                 FROM `bigquery-public-data.google_analytics_sample.ga_sessions_*`
                 WHERE _table_suffix BETWEEN '20170601' AND '20170630'
                 GROUP BY totals.transactions
                 HAVING totals.transactions >=1
                 )
                                        
            UNION ALL

            SELECT (SUM(pageviews)/SUM(users)) AS avg_number_of_product_pageviews
            FROM (
            SELECT sum(totals.pageviews) as pageviews,
            COUNT(fullvisitorid) as users
            FROM `bigquery-public-data.google_analytics_sample.ga_sessions_*`
            WHERE _table_suffix BETWEEN '20170701' AND '20170730'
            GROUP BY totals.transactions
            HAVING totals.transactions >=1
            )

            UNION ALL

            SELECT (SUM(pageviews)/SUM(users)) AS avg_number_of_product_pageviews
            FROM (
            SELECT sum(totals.pageviews) as pageviews,
            COUNT(fullvisitorid) as users
            FROM `bigquery-public-data.google_analytics_sample.ga_sessions_*`
            WHERE _table_suffix BETWEEN '20170801' AND '20170830'
            GROUP BY totals.transactions
            HAVING totals.transactions >=1
            ) '''

purchaseTrue_df = client.query(query).to_dataframe()
purchaseTrue_df

#What is the exit rate for each pageview in July 2017?
query ='''SELECT hits.page.pagePath,
                 count(*) as pageviews,
                 countif(hits.isEntrance) as entrances,
                 countif(hits.isExit) as exits,
                 round(100*(countif(hits.isExit)/count(*)), 2) as exitRate
                 FROM `bigquery-public-data.google_analytics_sample.ga_sessions_*`, 
                 unnest(hits) as hits 
                 WHERE hits.type = "PAGE"
                 AND _TABLE_SUFFIX BETWEEN '20170701' AND '20170731'
                 GROUP BY 1
                 ORDER BY 2 DESC'''
pageview = client.query(query).to_dataframe()
```
